```python
# IMPORTING THE NECESSARY HEADERS
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# LOADING THE DATASET

```python
columns=['Sepal Length','Sepal Width','Petal Length','Petal Width','Class Labels']
df=pd.read_csv('C:\\Users\\aksha\\Desktop\\DS course\\DATA\\iris.csv',names=columns)
```


# VISUALIZING THE DATASET

```python
columns=['Sepal Length','Sepal Width','Petal Length','Petal Width']
for x in columns:
    df[x]=pd.to_numeric(df[x])
sns.pairplot(df, hue='Class Labels')
```

# SEPARATING INPUT COLUMNS & OUTPUT COLUMNS 

```python
# Separate features and target variables

data=df.values
# data is a numpy array containing all the values of the dataframe df.
X= data[:,0:4]
Y= data[:,4]
print(Y)
```

# SPLITTING THE DATA INTO TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
print(Y_train)
```

# MODEL 1:
# Support Vector Machine Algorithm

```python
from sklearn.svm import SVC

model_svc=SVC()
model_svc.fit(X_train,Y_train)
```

```python
prediction1= model_svc.predict(X_test)
# Calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction1)*100)
```

# MODEL 2:
# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model_LR= LogisticRegression(max_iter=1000)
#Increase max_iter to allow more iterations
model_LR.fit(X_train,Y_train)
```

```python
prediction2=model_LR.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction2)*100)
```

```python
for i in range(len(prediction2)):
    print(Y_test[i],prediction2[i])
```

# MODEL 3: 
# DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
model_DTC=DecisionTreeClassifier()
model_DTC.fit(X_train,Y_train)
```

```python
prediction3=model_DTC.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction3)*100)
```

```python
# CLASSIFICATION REPORT
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction2))
```
